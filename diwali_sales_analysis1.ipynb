```python
#EXPLORATORY DATA ANALYSIS
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

```

```python
#FOR VALUE ON TOP
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount


sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ay=sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
for bars in ay.containers:
    ay.bar_label(bars)
```

```python
#In countplot show bifurcated Data of Male and Female 
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

# from above we can say that major buyers are female 

# ADVANCE ANALYSIS


```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

# this means 26-35 age group is most significant buyer

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

#sns.set(rc={'figure.figsize':(15,5)})
#to resize the state name
sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

# hence UP has most orders

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

#sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'Marital_Status',data = df)

sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot( x = 'Marital_Status',y= 'Amount', hue='Gender',data = sales_state)
```

# 0>>Married 1>>Unmarried

```python
sales_state=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})

sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_state)
```

# from above table observed that marriedFemale places more orders

# occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum()
#to give size of table/barplot
sns.set({'figure.figsize':(18,6)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

```python
# .sort_values(by='Amount',ascending=False used below for arrangig value high to low
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#to give size of table/barplot
sns.set({'figure.figsize':(18,6)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

# from above, most buyers are working in IT, HEALTHCARE,AVIATION etc.


```python
# product category vs count plot
sns.set(rc={'figure.figsize':(30,15)})
ax=sns.countplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#sort order domain>>food,clothes,electronic items
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

# which product has most no of orders acc to Product_ID

```python
# plotting barplot to show product sold vs order count data
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# Based on Product name
# top 5 most sold products (same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,12))
df.groupby('Product_Category')['Orders'].sum().nlargest(5).sort_values(ascending=False).plot(kind='bar')
```

# Conlusion:  Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Diwali Sales.


# The top 5 most Prominent Orders are from:  1>Clothing&Apparel 2>Food 3>Electronic Gadgets 4> Shoes 5>Household Items.
